## K Values

Here we compute the K values with the syntethic photometry previously obtained. The results are saved on /Data/K_values_OT

```python
%%capture
## Ahora hay que calcular el MAD Y MUTIPLICAR LOS ESPECTROS

# ## We calculate the syntethic photometry vs the observerd photometry, this in order to obtain the K values for each band and for each spectra

# Create output directories
try:
    os.mkdir(os.path.join(os.getcwd(),"Data", "K_values_OT"))
except FileExistsError:
    pass



list_sn=os.listdir(os.path.join(os.getcwd(),'Data','photometry_syn_OT_uncalibrated'))
all_rms_type=pd.Series([],dtype='float64')
#list_sn=['PTF12gzk_OT.dat']

for i in range(len(list_sn)):
    name=list_sn[i][:-18]

    #dir1=os.getcwd()+'/OT_combines_test/plot_syn_vs_obs/'+tipo #para los plot
    dir2=os.path.join(os.getcwd(),'Data','K_values_OT')#para guardar los valores de K por fase

    path=os.path.join(os.getcwd(),'Data','photometry_syn_OT_uncalibrated',str(name)+'_synphotometry.dat')
    path2=os.path.join(os.getcwd(),'Data','photometry_loess_syn',str(name)+'_photometry.dat')
    path_spec=os.path.join(os.getcwd(), "Data", "spectra_calibrated_1", str(name)+'_spectra.dat')
    #os.chdir(dir1)
    #import matplotlib.backends.backend_pdf
    #pdf = matplotlib.backends.backend_pdf.PdfPages(name+' syn_vs_obs.pdf')


    sintetica,filtros_sin=data_curvas(path)

    real,filtros_real=data_curvas(path2)

    spec,fases=leer_spec(path_spec,MJD=True,as_pandas=True) #this is just to plot the real fases after if you want


    all_rms=pd.Series([],dtype='float64')
    all_K_df=pd.DataFrame([])
    #filtros_sin=['g']
    for ii in range(len(filtros_sin)):
        index_sin=ii
        index_obs=filtros_real.index(filtros_sin[ii])


        real_df=real[index_obs]
        #real_df=real_df.loc[real_df.loc[:, 0] <100]
        sint_df=sintetica[index_sin]

        print(filtros_real,filtros_sin)
        #NOTAR QUE ACA ES PARA OT=TRUE, CUANDO OT=FALSE SE RECIVEN MENOS PARAMETROS
        try:
            rmss,mjd_merge,K,ini_merge,fin_merge,alpha_merge,ALR_df=sin_vs_obs(sint_df,filtros_sin[ii],real_df,
                                             K=True,ot=True,get_ALR=True,use_cte=False,mag_to_flux=False)
        except TypeError:
            #this is when the LC does not cover the spectra or viceversa
            continue
        a=[name]
        a=pd.Series(a*len(mjd_merge))
        K_df=pd.concat((a,mjd_merge, K,pd.Series(np.zeros(len(K))),rmss,
                        ini_merge,fin_merge,alpha_merge), axis=1,keys=['name','mjd','K','Band','residual','Ini','Fin','alpha'])
        #K_df.rename(columns={0:'mjd',0:'K',1:'band'},inplace=True)
        K_df.Band.replace([0], [filtros_sin[ii]], inplace=True)
        all_K_df=pd.concat((all_K_df,K_df),axis=0,ignore_index=True)

        #all_rms=all_rms.append(rmss,ignore_index=True)
        all_rms=pd.concat([all_rms, rmss], ignore_index=True)


        #pdf.savefig(fig,bbox_inches='tight')

    #all_rms_type=all_rms_type.append(all_rms,ignore_index=True)
    all_rms_type=pd.concat([all_rms_type,all_rms],ignore_index=True)
    #os.chdir(dir2)
    all_K_df=all_K_df.drop_duplicates()
    all_K_df.to_csv(os.path.join(str(dir2),str(name)+'_K.csv'), index=False)

```

## Rescaling

Here we calibrate the entire grid of spectra with the Mean K value obtained on the previous steps.  **Remember changue the values ABC on plano() in order to use that correspond with your SN type**. This cell create a file named rms_mad_K_OT with the information necesary to calibrate our spectra. This file is saved on /Data

```python
%%capture
dir1=os.path.join(os.getcwd(),'Data','K_values_OT')
list_sn=os.listdir(dir1)
#list_sn=["SN2007Y_K.csv"]


#Para calcular los rms y archivos de texto con la info del K y el epsilon
def plano(x,y):
    #for type II:       solution: 0.000769 x + 0.000229 y + 0.022254 = z
    #For type Ia/Ibc:   solution: 0.010329 x + -0.004641 y + 0.186775 = z
    A=0.010329
    B=-0.004641
    C=0.186775
    z=A*x+B*y+C
    return z


maximum_df=pd.read_csv(os.path.join(os.getcwd(),'Data','maximum_perband.dat'))





#os.chdir(os.getcwd()+'/OT_combines_test/')
file=open(os.path.join(os.getcwd(),'Data','rms_mad_K_OT.dat'),'w')
file.write('name,mjd,rms,mad,epsilon,median,K_values,bands,ini,fin,alpha\n')
print('name|mjd|rms|mad|epsilon|median|K_values|bands|ini|fin|alpha')


for elemento in list_sn:
    print(elemento)
    try:
        K_df=pd.read_csv(dir1+'/'+elemento)
    except:
        print(elemento,'is an empty file')
        continue

    mjd_unique=K_df['mjd'].unique()

    for i in range(len(mjd_unique)):

        a=K_df.loc[K_df.loc[:,'mjd']==mjd_unique[i]] #dataframe con las mismas fechas
        a_list=a['alpha'].unique()
        for j in range(len(a_list)):

            df_each=a.loc[a['alpha']==a_list[j]] #mismas fechas y alphas

            INI=list(df_each['Ini'])
            FIN=list(df_each['Fin'])
            for l in range(len(INI)):
                newdf=df_each.loc[((df_each['Ini']==INI[l] ))& ( df_each['Fin']==FIN[l])] #mismas fechas, alpahs y inicio -fin
                k_list=list(newdf['K'])
                bands=list(newdf['Band'])
                if len(k_list)>1:
                    name=K_df['name'][0]

                    k_promedio=np.mean(k_list)
                    k_mediano=np.median(k_list)

                    rms_k=rms(k_list,1)
                    mad_k=(L1(k_list)*1.5) #cuanto seria el numero de parametros aca?, el 1.5 es para comparar con el rms

                    #calculo el epsilon
                    dif_dias=float(FIN[l]-INI[l])
                    #if tipo =='Ia':
                    #    maximum=float(maximum_df.loc[(maximum_df['name']==name) & (maximum_df['filter']=='V' )]['mjd_max'])
                    #if tipo =='II' or tipo =='Ib' or tipo =='Ic':
                    #    print(name)
                    #    print((maximum_df.loc[(maximum_df['name']==name)]['fase']))
                    #    maximum=float(maximum_df.loc[(maximum_df['name']==name)]['fase'])
                    band_system=['V','R','r','i','I','g','B','U','u','z'] #We use this order to obtain the day of the maximum
                    for elemento in band_system:

                        maximum=maximum_df.loc[(maximum_df['name']==name) & (maximum_df['filter']==elemento )]['mjd_max']
                        if len(maximum)>0:
                            maximum=float(maximum)
                            break
                    phase=mjd_unique[i]-maximum

                    epsilon=float(plano(dif_dias,phase))

                    #epsilon=recta(dif_dias)
                    for b in range(len(k_list)):
                        k_list[b]=str('%.5f'%k_list[b])

                    print(name,mjd_unique[i],rms_k,mad_k,k_mediano,bands,epsilon)
                    file.write(name+','+str(mjd_unique[i])+','+str(rms_k)+','+str(mad_k)+','+str(epsilon)+','+str(k_mediano)+
                               ','+str('-'.join(k_list))+','+
                               str(''.join(bands))+','+str(INI[l])+','+str(FIN[l])+','+str(a_list[j])+'\n')
                else:
                    print(k_list,'muy pocos datos para calcular rms')
                    K_df.loc[K_df.loc[:,'mjd']==mjd_unique[i]]

file.close()

# Delete duplicated
rms_mad_df=pd.read_csv(os.path.join(os.getcwd(),'Data','rms_mad_K_OT.dat'))
rms_mad_df.drop_duplicates(inplace=True)
rms_mad_df.to_csv(os.path.join(os.getcwd(),'Data','rms_mad_K_OT.dat'),index=False)


```

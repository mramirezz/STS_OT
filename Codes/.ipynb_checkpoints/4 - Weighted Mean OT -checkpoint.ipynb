# Wighted Mean

In this section, we will calculate the weighted mean of the spectra. As input, we have all possible combinations of spectra calculated in code 3 and already calibrated (stored in Data/OT_calibrated_1). The spectra are weighted based on their  date and the relative error of their calibration (MAD). Additionally, edge corrections are applied as not all spectra have the same length. For further details, please refer to the work by Ramirez M (2023).

```python
%%capture
# We read the mad file
rms_mad_df=pd.read_csv(os.path.join(os.getcwd(),'Data','rms_mad_K_OT.dat'))
# Create output directories
try:    
    os.mkdir(os.path.join(os.getcwd(),"Data", "spectral_time_series"))
except FileExistsError:
    pass

try:    
    os.mkdir(os.path.join(os.getcwd(),"Data", "spectral_time_series_prom_plots"))
except FileExistsError:
    pass



#path_sns="G:/Mi unidad/Work/OT_unidos_2/OT_combines_test/II_reescalados_separated/"
path_sns=os.path.join(os.getcwd(),'Data','OT_calibrated_1')
supernovas=os.listdir(path_sns)
#supernovas=['SN2007Y_OT.dat']


#leo el archivo con las MAD que voy a utilizar
from time import time
start_time = time()

ventana=50 #window for the edges
for sn in supernovas:
    ESPECTRO,fases,ini,fin,alpha=leer_spectra_OT(os.path.join(path_sns,sn))

    name=sn[:-7]

    print('Entrando a;',name)

    file1=open(os.path.join(os.getcwd(),'Data','spectral_time_series',str(name)+'.dat'),'w')
    
    import matplotlib.backends.backend_pdf
    pdf = matplotlib.backends.backend_pdf.PdfPages(os.path.join(os.getcwd(),"Data", "spectral_time_series_prom_plots",str(name)+'.pdf'))

    
    fases=[int(x) for x in fases]
    

    from collections import defaultdict
    Ubicaciones = defaultdict(list)
    for index in range(len(fases)):
        Ubicaciones[fases[index]].append(index)
    Ubicaciones

    fases_unique=pd.Series(fases).unique()
    fases_unique=sorted(fases_unique)
                                                   
                                                   
    from functools import reduce
    for j in range(len(fases_unique)):

        #if float(fases_unique[j])<=56268  :
            #print('fase continuamos:',fases_unique[j])
            #continue

        spec_posiciones=Ubicaciones[fases_unique[j]] #posiciones de los espectros con la misma fase 
        #el 0 en fases_unique debe iterar para tener todos los espectros
        lista_ESPECTRO=[]
        lista_iniciales=[]
        lista_finales=[]
        lista_alphas=[]
        for posicion in spec_posiciones:
            #print(posicion)
            #ESPECTRO,fases2,ini,fin,alpha=leer_spec(path_sns+'/'+name+'/'+fases_sn[posicion],ot=True)
            lista_ESPECTRO.append(pd.DataFrame(ESPECTRO[posicion]))
            lista_iniciales.append(float(ini[posicion]))
            lista_finales.append(float(fin[posicion]))
            lista_alphas.append(float(alpha[posicion]))

        lista_merge=[]
        #generamos la lista con los valores de C
        C_list=[]
        for p in range(len(lista_ESPECTRO)):
            wave_array=np.array(lista_ESPECTRO[p]['wave'])
            c=np.ones(len(lista_ESPECTRO[p]))
            c[:ventana]=np.linspace(0,1,ventana)
            c[-ventana:]=np.linspace(1,0,ventana)
            #plt.plot(wave_array,c)
            #plt.show()
            df_c=pd.DataFrame({'wave':wave_array,'c':c})
            C_list.append(df_c) #es una lista con los df de C para cada espectro y longitud de onda
        #ahora tengo que recorrer todas las longitudes de onda para hacer el promedio para esta fase
            merge=pd.merge(lista_ESPECTRO[p],df_c,on='wave')
            lista_merge.append(merge)

        lista_MAD_eps=[]
        lista_sigma_cuadrado=[]
        for l in range(len(lista_ESPECTRO)): #obtengo los MAD y epsilon de los espectros
            df_=rms_mad_df.loc[(rms_mad_df['name']==name)&(rms_mad_df['mjd']==fases_unique[j]) & 
                   (rms_mad_df['ini']==lista_iniciales[l]) & 
                   (rms_mad_df['fin']==lista_finales[l])] #este es el df con los datos de cada espectro
            #print(df_)
            df_=df_.loc[df_['name']==name] #para el bug cuando salen repetidos
            MAD=df_['mad']                                #obtengo el mad del espectro
            MAD=float(MAD)/1.5 #divido por 1.5, porque lo habia mutiplicado por 1.5 para comparar con rms en la golden sample
            #dif_dias=float(df_['fin']-df_['ini'])
            #epsilon=recta(dif_dias)                       #obtengo el epsilon del espectro
            epsilon=df_['epsilon']
            epsilon=float(epsilon)

            if epsilon <0.01:
                epsilon=0.01
            #print('------------')
            #print('MAD:',MAD)
            #print('Epsilon:',epsilon)

            #print('------------')
            lista_MAD_eps.append([MAD,epsilon])
            lista_sigma_cuadrado.append((MAD**2)+(epsilon**2))

        for indice in range(len(lista_ESPECTRO)):
            MAD=lista_MAD_eps[indice][0]
            epsilon=lista_MAD_eps[indice][1]

            lista_merge[indice]['numerador']= (lista_merge[indice]['flux'].mul(lista_merge[indice]['c']))/((MAD**2)+(epsilon**2))

            lista_merge[indice]['denominador']=(1*lista_merge[indice]['c'])/((MAD**2)+(epsilon**2))


            lista_merge[indice].set_index('wave', inplace = True)


        prueba=reduce(lambda a, b: a.add(b, fill_value=0), lista_merge)

        prueba['division']=prueba['numerador']/prueba['denominador']

        #plt.plot(prueba.index,prueba['division'])
        df_spec_prom=pd.DataFrame({'wave':np.array(prueba.index),'flux':prueba['division']}).dropna().reset_index(drop=True)

        
        fig = plt.figure(figsize=(15,12))
        fig.subplots_adjust(left=0.12, bottom=0.2, hspace=0, right=0.99, top=0.96)
        ax = fig.add_subplot(211)
        ax_ = fig.add_subplot(212)
        plt.title(fases_unique[j])

        #fig=plt.figure(figsize=(12,8))
        peso=1/np.array(lista_sigma_cuadrado[:len(lista_ESPECTRO)])
        peso=peso/np.max(peso)
        peso_sorted=sorted(peso)
        
        import matplotlib.cm as cm
        cmap = plt.get_cmap('rainbow', len(peso))
        colors = cm.rainbow(np.linspace(0,1,len(peso)))
        
        for ll in range(len(lista_ESPECTRO)):
            merge2 = pd.merge(df_spec_prom, lista_ESPECTRO[ll], how='inner', on=['wave'])
            #ax.plot(np.array(lista_ESPECTRO[ll]['wave']),np.array(lista_ESPECTRO[ll]['flux']),label='%.3f'%peso[ll]
            #         +', '+'%.3f'%lista_MAD_eps[ll][0]+', '+'%.3f'%lista_MAD_eps[ll][1],color=colors[peso_sorted.index(peso[ll])])
            ax.plot(np.array(lista_ESPECTRO[ll]['wave']),np.array(lista_ESPECTRO[ll]['flux']),label=str(lista_iniciales[ll])+','+str(lista_finales[ll]) ,color=colors[peso_sorted.index(peso[ll])])
            residuales_pesados=((merge2['flux_x']-merge2['flux_y'])/((merge2['flux_x']+merge2['flux_y'])/2))*peso[ll]
            ax_.plot(np.array(merge2['wave']),np.array(residuales_pesados),color=colors[peso_sorted.index(peso[ll])])

        ax_.axhline(0,ls='--',color='black',alpha=0.5)
        ax.plot(np.array(df_spec_prom['wave']),np.array(df_spec_prom['flux']),'--',label='prom',color='black')
        ax.legend(bbox_to_anchor=(0,1.1), loc="lower left",ncol=5,fontsize='large')
        ax.set_ylabel('Flux')
        ax_.set_xlabel('Wave')
        ax_.set_ylabel('Residuals*weight')

        # Normalizer
        import matplotlib as mpl
        norm = mpl.colors.Normalize(vmin=0, vmax=1)

        # creating ScalarMappable
        sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
        sm.set_array([])
        # add color bar below chart
        from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
        ax_divider = make_axes_locatable(ax)
        # define size and padding of axes for colorbar
        cax = ax_divider.append_axes('top', size = '5%', pad = '8%')
        plt.colorbar(sm,cax=cax,orientation='horizontal',ticks=[0,1])
        plt.tight_layout()

        
        pdf.savefig(fig)
        plt.show()
        



        
        #ahora escribimos el espectro promediado
        file1=open(os.path.join(os.getcwd(),'Data','spectral_time_series',str(name)+'.dat'),'a')

        file1.write('# time:\t'+str(fases_unique[j])+'\n')
        file1.write('# SPEC \n')
        file1.write('#      WAVE   FLUX')
        for ii in range(len(df_spec_prom)):
            file1.write('\n')
            file1.write(str(df_spec_prom.iloc[ii]['wave'])+'\t'+str(float(df_spec_prom.iloc[ii]['flux'])))
        file1.write('\n')
        file1.close()

        


        print('fase done:',fases_unique[j],name)   
        print('###################################################')
        #limpiamos memoria
        del fig
        del lista_ESPECTRO
        del lista_iniciales
        del lista_finales
        del lista_alphas
        del df_spec_prom
        del prueba #es el spec_prom
        del C_list
        del merge
        del merge2
        #gc.collect()

        #break
    file1.close()
    pdf.close()
    
elapsed_time = time() - start_time
```

```python
#we look for the file with the day of the maximum, to put plots in phases, if the user have a different file, try to match the
#format on the file /Data/maximum_perband.dat


def lc_max_fromfile(name): 
    #input the name of the SN as as string
    #search on the datafram of maximum for the day of the maximum and return the value
    #search for differents bands available
    maximum_df=pd.read_csv(os.path.join(os.getcwd(),'Data','maximum_perband.dat'))

    band_system=['V','R','r','i','I','g','B','U','u','z'] #We use this priority order to determine the day with the maximum value.

    for elemento in band_system:

        maximum=maximum_df.loc[(maximum_df['name']==name) & (maximum_df['filter']==elemento )]['mjd_max']
        if len(maximum)>0:
            maximum=float(maximum)
            break
    return maximum

```

```python
a = fases
b = fases2
a_data = ESPECTRO
b_data = ESPECTRO2


# Combinar y ordenar las listas
combined = [(num, data, 'a') for num, data in zip(a, a_data)] + [(num, data, 'b') for num, data in zip(b, b_data)]
combined_sorted = sorted(combined, key=lambda x: x[0])
lista_ordenada = [x[0] for x in combined_sorted]
lista_data_ordenada = [x[1] for x in combined_sorted]
lista_origen = [x[2] for x in combined_sorted]

# Calcular la diferencia máxima entre los valores de flujo de cada espectro consecutivo
max_diff = 0
for i in range(len(lista_data_ordenada) - 1):
    diff = np.max(np.abs(lista_data_ordenada[i + 1]['flux'] - lista_data_ordenada[i]['flux']))
    if diff > max_diff:
        max_diff = diff

# Utilizar un porcentaje de la diferencia máxima como el valor de cte
cte_factor = 0.25
cte_step = cte_factor * max_diff

# Graficar los espectros
fig = plt.figure(figsize=(12, 9))

cte = 0

for i in range(len(lista_ordenada)):
    if lista_origen[i] == 'a':
        color = 'orange'
    else:
        color = 'C0'

    x_data = lista_data_ordenada[i]['wave']
    y_data = lista_data_ordenada[i]['flux'] - cte
    plt.plot(x_data, y_data, color=color)

    x_text = x_data.iloc[-1] - 0
    y_text = y_data.iloc[-1]
    plt.text(x_text, y_text, f": {lista_ordenada[i]}", fontsize=10, color='black')

    cte = cte + cte_step

plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('Espectros Ordenados')
plt.show()
```

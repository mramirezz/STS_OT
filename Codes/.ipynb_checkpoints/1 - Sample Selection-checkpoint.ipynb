The goal of this notebook is to carefully curate a subset of high-quality supernova data that will be used to construct a reliable spectral time series. As a selection criterion, we will use the comparison between synthetic and observed photometry. Specifically, we will calculate the ratio $K=\frac{F_{obs}}{F_{syn}}$, where $F_{obs}$ is the observed flux and $F_{syn}$ is the corresponding synthetic flux. This approach allows us to select only the highest quality observations for use in our analysis.

In addition, we will use the median absolute deviation (MAD) to evaluate the quality of each $K$ value. This metric will be applied to all $K$ values obtained for each spectral band covered by the supernova light curve. By using the MAD, we can assess the variability of the $K$ values and identify and remove any outliers that may be present in our dataset.

We now compute the splines and store the corresponding text files and images in Data/spectra_spline and Data/spectra_spline_images, respectively.

```python
#To not overload with plot this cell

%matplotlib agg
from scipy.ndimage import median_filter


from scipy import interpolate
from scipy.ndimage import median_filter

import matplotlib.backends.backend_pdf


#spectra_list=['SN2007bm_spectra.dat']
#inicio=spectra_list.index('SN2012hn_spectra.dat') #this is when we want to do an specific SN


inicio=0
path_spec = "/Data"
# Create output directories
try:
    os.mkdir(os.path.join(os.getcwd(),"Data", "spectra_splines"))
    os.mkdir(os.path.join(os.getcwd(),"Data", "spectra_splines_images"))
except FileExistsError:
    pass
print("Files saved on", path_spec+"/spectra_splines")
print("Images saved on", path_spec+"/spectra_splines_images")

for index in range(inicio,len(spectra_list)):

    #we read the spectra file
    sn=spectra_list[index]

    ESPECTRO,fases=leer_spec(os.path.join(os.getcwd(),"Data", "spectra",str(sn)),MJD=True,as_pandas=True)
    print(fases)
    print("Spline on: ",index+1,sn)

    path_splines=os.path.join(os.getcwd(),"Data", "spectra_splines")
    path_splines_images=os.path.join(os.getcwd(),"Data", "spectra_splines_images")


    #os.chdir(path_splines)
    name=spectra_list[index][:-12] #es por el nombre de los archivos, quitamos el _spectra.dat

    if len(fases)>2: #todas deberian ser mayor que 2
        #---Calculamos el spline y lo guardamos

        txt_spec_spline = open(path_splines+"/"+str(name)+'_spline.dat','w') #archivo de texto donde se guardan los spline

        #os.chdir(path_splines_images)



        import matplotlib.backends.backend_pdf
        pdf = matplotlib.backends.backend_pdf.PdfPages(path_splines_images+'/'+name+'_splines.pdf')


        for j in range(len(ESPECTRO)):


                ESPECTRO[j] = ESPECTRO[j].dropna()
                ESPECTRO[j].drop_duplicates(subset=['wave'], keep='first', inplace=True)



                WAVE_SPEC=np.array(ESPECTRO[j]['wave']) #longitud de onda del spectro [j]
                FLUX_SPEC=np.array(ESPECTRO[j]['flux']) #flujo del spectro[j]

                #Hacemos el smooth

                # Define el tamaño de la ventana de mediana en unidades de longitud
                #window_size = 5.0  # en Angstrom

                # Aplica el filtro de mediana
                #flux_smooth = median_filter(FLUX_SPEC, size=int(window_size/np.mean(np.diff(WAVE_SPEC))))
                #ax = fig.add_axes([0,0,1,1])

                ####### AHORA HACEMOS EL SPLINE
                from scipy import interpolate
                cs = interpolate.interp1d(WAVE_SPEC, FLUX_SPEC,fill_value = "extrapolate")
                #cs = CubicSpline(WAVE_SPEC, FLUX_SPEC)
                xnew=np.arange(min(WAVE_SPEC),max(WAVE_SPEC)+1,1)
                diferencia=(xnew[0]-int(xnew[0]))

                xnew=xnew-diferencia

                #LO ESCRIBIMOS EN EL ARCHIVO DE TEXTO
                txt_spec_spline.write('#\tSPEC\t-\tat\tMJD='+str(fases[j])+'\n')
                txt_spec_spline.write('#     WAVE     FLUX\n')
                for k in range(len(xnew)):
                    txt_spec_spline.write('     '+'%.2f'%xnew[k]+'        '+'%.5e'%cs(xnew[k])+'\n') #lo guardamos con el formato deseado


                #Plot
                fig2 = plt.figure(figsize=(10,5))
                ax= fig2.add_subplot(111)


                ax.plot(WAVE_SPEC,FLUX_SPEC,label='data') #ploteamos el espectro de los datos
                ax.plot(xnew,cs(xnew),'-',color='orange',label='spline',alpha=0.5) #ploteamos el spline
                #ax.text(1, 1, 'matplotlib', horizontalalignment='right',verticalalignment='top', transform=ax.transAxes)
                #plt.plot(xpred,ypred,'-',label='GP')
                plt.legend()
                ax.set_xlabel('WAVE')
                ax.set_ylabel('FLUX')
                plt.title(name+' MJD='+str(fases[j]))



                plt.legend()
                #plt.show()

                pdf.savefig(fig2)
                fig2.clf()  # limpia la figura para comenzar de nuevo


        pdf.close()
        #os.chdir(path_splines)
```

After obtaining the Synthetic photometry for our spectra, it is imperative to compute the K value. This parameter plays a crucial role in the calibration of our spectra, as it enables us to select those that are suitably calibrated.

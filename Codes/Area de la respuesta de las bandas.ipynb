#### Ploteamos las bandas para ver como se ven con su respectivo ajuste spline, tambien guardamos los datos del ajuste en un archivo de texto.

```python
nombre_carpeta=input('Ingrese nombre carpeta (#Splines_response para slowpe)') #Splines_response para slowpe

if os.path.exists(nombre_carpeta):
    shutil.rmtree(nombre_carpeta) #nos aseguramos que no exista la carpeta

os.mkdir(nombre_carpeta) #creamos la carpeta donde iran guardados estos archivos y figuras.

areas=[]


for j in range(len(datos)):
    archivo1 = open('spline_'+str(bandas[j]),'w') #abrimos un archivo de texto por filtro que vamos a escribir
    #archivo1.write('\n')
    #archivo1.write('\n')
    #archivo1.write('\n')
    #archivo1.write('\n')
    if select==2:
        f1=open (os.getcwd()+'/filters/'+str(bandas[j])+'.txt','r+') #abrimos cada archivo antiguo
    elif select==1:
        f1=open (os.getcwd()+'/'+str(bessell)+'/'+str(bandas[j]),'r+')

    
    
    #para guardar la cabezera y el formato de los archivos de texto antiguo
    cabezera=f1.readlines() #leimos el archivo completo

            
    wave=[]
    response=[]
    
    if select==1:
        for i in range(len(datos[j])):
            if i<=3:
                archivo1.write(cabezera[i]) #escribimos solo la cabezera
                           
            if i>3: #para empezar a guardar desde los datos que necesito
                if float(datos[j][i][0])!=0:
                    wave.append(np.float(datos[j][i][0]))
                    if datos[j][2][5]=='1': #quiere decir q esta en energia y hay que dividir por lambda la respuesta
                        print('Pasamos a photones')
                        response.append(np.float(datos[j][i][1])/np.float(datos[j][i][0]))
                    else:    
                        response.append(np.float(datos[j][i][1])) #usamos la lista de datos[] para trabajar con los datos
                        #creamos la lista de datos de wave y respuesta
    
    
    
    elif select==2:
        if lista[j][0]=='s':
            for a in range(len(datos[j])): #con este guarda los archivos en la lista 
                if a>0:                
                    wave.append(np.float(datos[j][a][0]))
                    response.append(np.float(datos[j][a][1]))
                    
        elif lista[j][8]!='p':
            for a in range(len(datos[j])): #con este guarda los archivos en la lista 
                if a>3:                
                    wave.append(np.float(datos[j][a][0]))
                    response.append(np.float(datos[j][a][1]))
        elif lista[j][8]=='p':
            #continue
            for a in range(len(datos[j])): #con este guarda los archivos en la lista 
                if a>2:                
                    wave.append(np.float(datos[j][a][0]))
                    response.append(np.float(datos[j][a][1]))
                    
            
    #para eliminar los valores repetidos del wave------------------
    lista_original=wave
    lista_original2=response
    wave=[]
    response=[]
    for i in range(len(lista_original)):
         if lista_original[i] not in wave:
            wave.append(lista_original[i])
            response.append(lista_original2[i])
    #---------------------------------------------------------------
            
    xnew=np.arange(int(min(wave)),max(wave)+1,1) #arreglo en los cuales evaluare el spline, debe tener el mismo rango que wave, es la grilla
    cs = CubicSpline(wave, response) #aplicando la funcion spline en mi respuesta
    

    
    for k in range(len(xnew)):
        
        archivo1.write('     '+str(xnew[k])+'        '+str(cs(xnew[k]))+'\n') #lo guardamos con el formato deseado
    
    
    areatotal=0
    for l in range(len(xnew)-1):
        b1=xnew[l]
        b2=xnew[l+1]
        base=b2-b1
        h1=cs(xnew[l])
        h2=cs(xnew[l+1])
        prom=(h1+h2)/2
        area=base*prom
        areatotal=area+areatotal
    
    areas.append(areatotal) 
    
    shutil.move('spline_'+str(bandas[j]),nombre_carpeta)
    
    print('el numero de datos es ',len(xnew))
    
    fig=plt.figure()
    plt.plot(xnew,cs(xnew),color='orange',label='spline')
    plt.scatter(wave,response,s=15,color='b',label='data')    
    plt.xlabel('WAVE [amstrong]')
    plt.ylabel('RESPONSE')
    plt.title('RESPUESTA DE LA BANDA '+str(bandas[j]))
    plt.legend()
    plt.show()
    figname=str(bandas[j]).replace('.dat','')
    figname=str(bandas[j]).replace('.txt','')
    fig.savefig(figname+'.png', bbox_inches='tight')
    shutil.move(figname+'.png',nombre_carpeta)


       
```
